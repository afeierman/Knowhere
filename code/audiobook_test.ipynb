```python
import pymongo
import pandas as pd
import re
import dateutil
foobar = re.compile('(?P<h>\d*)(?= hrs)?\D*(?P<m>\d*)(?= mins)?')
```

```python
def get_total_time(time):
    try:
        time = re.match(pattern=foobar,string=time)
        td = time.groupdict()
        td = {k: int(v) if v is not '' else 0 for k,v in td.iteritems()}
        total_mins = td['h'] * 60 + td['m']
        total_mins = total_mins / 15 * 15
    except:
        return 0
    return total_mins

def get_release_date(date):
    try:
        date = dateutil.parser.parse(date)
    except:
        return pd.np.nan
    return date
```

```python
df = pd.read_csv(r'C:/Users/William/Downloads/Audible.csv')
```

```python
df['Length'] = df['Length'].apply(lambda t: get_total_time(t))
df['RelDate'] = df['RelDate'].apply(lambda d: get_release_date(d))
df = df[df['OverallRating'] > 0]
df = df[df['Length'] <= 30 * 60]
df = df[df['Length'] >= 60]
df = df[pd.notnull(df['RelDate'])]
df = df[df['NumOverRating'] >= 10]
```
