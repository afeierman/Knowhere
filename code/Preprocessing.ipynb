```python
import db.knowhere_db as kdb
import pipeline.pipeline as pipeline
import pandas as pd
import numpy as np
```

```python
reader = kdb.Reader('knowhere')
```

```python
emil_H_data_raw = reader.get_dataframe_pivoted(collection='iphone_test2', username='emil', min_date="2017-03-19",
                                              max_date="2017-03-20")
```

```python
print len(emil_H_data_raw)
emil_H_data_raw.columns
```

```python
emil_H_data = emil_H_data_raw[['Acceleration (via User) (x)','Acceleration (via User) (y)',\
                               'Acceleration (via User) (z)', 'Altimeter (Barometer) (Pressure)',\
                           'Microphone (Left Channel Level)', 'Microphone (Right Channel Level)',\
                               'Magnetometer (raw) (x)', 'Magnetometer (raw) (y)','Magnetometer (raw) (z)',\
                               'Gyrometer (smooth) (x)','Gyrometer (smooth) (y)', 'Gyrometer (smooth) (z)']]

emil_H_data = emil_H_data.dropna().astype(float)
```

```python
emil_H_data['Acceleration'] =  np.sqrt(emil_H_data['Acceleration (via User) (x)']**2 +\
                                       emil_H_data['Acceleration (via User) (y)']**2 +\
                                       emil_H_data['Acceleration (via User) (z)']**2)
emil_H_data['Magnetometer'] =  np.sqrt(emil_H_data['Magnetometer (raw) (x)']**2 + emil_H_data['Magnetometer (raw) (y)']**2 +\
                                       emil_H_data['Magnetometer (raw) (z)']**2)
emil_H_data['Gyrometer']    =  np.sqrt(emil_H_data['Gyrometer (smooth) (x)']**2 + emil_H_data['Gyrometer (smooth) (y)']**2 +\
                                       emil_H_data['Gyrometer (smooth) (z)']**2)
emil_H_data['Microphone']    =  (emil_H_data['Microphone (Left Channel Level)'] +\
                                emil_H_data['Microphone (Right Channel Level)'])/2
emil_H_data = emil_H_data[['Acceleration','Magnetometer','Gyrometer','Microphone','Altimeter (Barometer) (Pressure)']]
```

```python
emil_H_data = emil_H_data.resample("2S").mean()
```
